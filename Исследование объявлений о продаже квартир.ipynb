```python
# Использую метод isna() и смотрю сумму пропусков по столбцам.
data.isna().sum()
```

Обработаем пропущенные значения по каждому параметру в отдельности.


### Пропуски в столбце "Аппартаменты"

Заменим пустые значение на логическое отрицание. Несмотря на возможность умышленного пропуска, большинство пропусков все же можно отнести на человеческий фактор.

```python
# Использую метод fillna() к пустым значениям в столбце "is_apartment" и делаю проверку
data['is_apartment'] = data['is_apartment'].fillna(False)
data['is_apartment'].isna().sum()
```

### Пропуски в количестве балконов
Заменяю все строчки с пропущенным значением балкона на "0". Предполагаю, что для помещений без балконов ставили в соответсвующем столбце пропуск.

```python
# Использую метод fillna() к пустым значениям в столбце "balcony" и делаю проверку
data['balcony'].fillna(0, inplace=True)
data['balcony'].isna().sum()
```

### Пропуски в указании названия населенного пункта

Данные о местоположении квартиры являются ключевым параметром влияющим на стоимость квартиры. Строки без указания населенного пункта удалим, так как нет адекватной замены этому признаку. Восстановить эти ячейки этого столбца по имеющимся данным нет возможности. 

```python
# С помощью логического оператора НЕ (~) и метода isna(), удалем из датафрейма пустые значение столбца locality_name
data = data[~data['locality_name'].isna()]
data['locality_name'].isna().sum()
```

```python
# Проверяю населенные пункты на уникальность. Ищу неявные дубликаты
data['locality_name'].unique()
```

```python
data['locality_name'].value_counts()
```

В названии населенных пунктов в одной основе слова используются разные буквы "е" и "ё". Это может быть причиной неявных дублей.
Заменим "ё" на "е".

```python
data['locality_name'] = data['locality_name'].replace('посёлок', 'поселок', regex=True)
data['locality_name'].unique()                                                                              
```

### Расстояние до ближайшего парка и расстояние до ближайшего водоёма

Данные до ближайшего парка и водоема определяет система. Данные возможно утеряны при выгрузке или система не смогла их расчитать ввиду того что нет парков и водоемов удовлетворяющих условиям "ближайший".

### Изучение факторов влияющих на цену квартиры

Выявляю зависимость цены от площади квартиры

```python
def func_scatter(df=None, x=None, y=None, alpha=None):
    plt.figure(figsize=(10,4))
    plt.title('Зависимость '+y+' от '+x+'', fontsize=16)
    plt.xlabel(x, fontsize=14)
    plt.ylabel(y, fontsize=14)
    plt.scatter(df[x], df[y], alpha=alpha)
    plt.grid(True)
    plt.show()
```

```python
def func_pivot(df):
    plt.figure(figsize=(10,4))
    plt.title('Зависимость '+df.columns[0]+' от '+df.index.name+'', fontsize=16)
    plt.xlabel(df.index.name, fontsize=14)
    plt.ylabel(df.columns[0], fontsize=14)
    ax = plt.axes()
    ax.plot(df.index, df.values)
    plt.grid(True)
    plt.show()
```

```python
func_scatter(data, 'total_area', 'last_price')
print('Коэффициент корреляции общей площади и цены', data['total_area'].corr(data['last_price']))
```

Есть положительная зависимость цены квартиры от её площади, хотя есть и выбивающиеся из этой зависимости объявления.

**Зависимость цены от числа комнат**

```python
func_scatter(data, 'rooms', 'last_price')
print('Коэффициент корреляции количества комнат и цены', data['rooms'].corr(data['last_price']))
```

Зависимость цены от количества комнат есть, но не сильная. График не наглядный. Сгруппируем данные по числу комнат.

```python
func_pivot(
    df = data.pivot_table(index='rooms', values='last_price', aggfunc='median')
)
```

Зависимость цены от количества комнат положительная. У квартир более четырех комнат цена начинает подниматься резко вверх, видимо связано с премиальных сегметом недвижимости.  Слабая корреляция цены от количества комнат может быть связана с ремонтом и обстановкой квартиры, принадлежности к престижному району.

**Зависимость цены от удалённости от центра**
